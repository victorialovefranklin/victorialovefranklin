```python
import requests  # Import the requests library

#Title: SARS-CoV-2 Variant Proportions

# Query URL
url =  ('https://data.cdc.gov/resource/jr58-6ysp.json')
print(url)

response = requests.get(url)  # Make a GET request to the URL

# Print status code (and associated text)
print(f"Request returned {response.status_code} : '{response.reason}'")

# Print data returned (parsing as JSON)
sarsvariantproportions= response.json()  # Parse `response.text` into JSON

import pprint
ppt = pprint.PrettyPrinter(indent=1)
ppt.pprint(sarsvariantproportions)
```
