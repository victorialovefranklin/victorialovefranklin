```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import ccf
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from datetime import timedelta

# Generate sample time series data
np.random.seed(42)
dates = pd.date_range(start="2022-01-01", periods=100, freq="D")
wastewater_viral_load = np.random.normal(loc=100, scale=10, size=100).cumsum()
clinical_cases = np.roll(wastewater_viral_load, 7) + np.random.normal(loc=0, scale=50, size=100)  # lagged + noise

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Wastewater_Viral_Load": wastewater_viral_load,
    "Clinical_Cases": clinical_cases
})
df.set_index("Date", inplace=True)

# Calculate Pearson and Spearman correlations
pearson_corr, _ = pearsonr(df["Wastewater_Viral_Load"], df["Clinical_Cases"])
spearman_corr, _ = spearmanr(df["Wastewater_Viral_Load"], df["Clinical_Cases"])

# Lag correlation function
lags = range(0, 15)
lagged_corrs = [pearsonr(df["Wastewater_Viral_Load"].shift(lag).dropna(),
                         df["Clinical_Cases"][lag:])[0] for lag in lags]

# Fit linear regression
X = df["Wastewater_Viral_Load"].values.reshape(-1, 1)
y = df["Clinical_Cases"].values
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

# Evaluate model
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(12, 14))

# Time series plot
axs[0].plot(df.index, df["Wastewater_Viral_Load"], label="Wastewater Viral Load")
axs[0].plot(df.index, df["Clinical_Cases"], label="Clinical Cases")
axs[0].set_title("Wastewater Viral Load vs. Clinical COVID-19 Cases Over Time")
axs[0].legend()

# Lag correlation plot
axs[1].bar(lags, lagged_corrs)
axs[1].set_title("Lagged Pearson Correlation (Wastewater vs. Clinical Cases)")
axs[1].set_xlabel("Lag (days)")
axs[1].set_ylabel("Correlation Coefficient")

# Regression plot
axs[2].scatter(df["Wastewater_Viral_Load"], df["Clinical_Cases"], alpha=0.6)
axs[2].plot(df["Wastewater_Viral_Load"], y_pred, color="red", label=f"Linear Fit\nRMSE: {rmse:.2f}, R: {r2:.2f}")
axs[2].set_title("Linear Regression: Predicting Clinical Cases from Wastewater Viral Load")
axs[2].set_xlabel("Wastewater Viral Load")
axs[2].set_ylabel("Clinical Cases")
axs[2].legend()

plt.tight_layout()
plt.show(), pearson_corr, spearman_corr
```

```python
# Extract years from the date index
df['Year'] = df.index.year

# List of unique years in data
years = df['Year'].unique()
```

```python
def analyze_yearly_data(df_year):
    # Calculate Pearson and Spearman correlations
    pearson_corr, _ = pearsonr(df_year["Wastewater_Viral_Load"], df_year["Clinical_Cases"])
    spearman_corr, _ = spearmanr(df_year["Wastewater_Viral_Load"], df_year["Clinical_Cases"])

    # Lag correlation for lags 0-14 days
    lags = range(0, 15)
    lagged_corrs = []
    for lag in lags:
        shifted_wastewater = df_year["Wastewater_Viral_Load"].shift(lag).dropna()
        clinical_aligned = df_year["Clinical_Cases"][lag:]
        corr = pearsonr(shifted_wastewater, clinical_aligned)[0]
        lagged_corrs.append(corr)

    # Linear Regression
    X = df_year["Wastewater_Viral_Load"].values.reshape(-1, 1)
    y = df_year["Clinical_Cases"].values
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)

    rmse = np.sqrt(mean_squared_error(y, y_pred))
    r2 = r2_score(y, y_pred)

    return {
        "pearson_corr": pearson_corr,
        "spearman_corr": spearman_corr,
        "lagged_corrs": lagged_corrs,
        "model_rmse": rmse,
        "model_r2": r2,
        "model": model
    }
```

```python
yearly_results = {}

for year in years:
    df_year = df[df['Year'] == year]
    results = analyze_yearly_data(df_year)
    yearly_results[year] = results

    print(f"Year: {year}")
    print(f"Pearson correlation: {results['pearson_corr']:.3f}")
    print(f"Spearman correlation: {results['spearman_corr']:.3f}")
    print(f"Lagged correlations (0-14 days): {[round(c, 3) for c in results['lagged_corrs']]}")
    print(f"Linear Regression RMSE: {results['model_rmse']:.2f}")
    print(f"Linear Regression R^2: {results['model_r2']:.3f}")
    print("-" * 40)
```

```python
fig, axs = plt.subplots(len(years), 1, figsize=(12, 5 * len(years)))

if len(years) == 1:
    axs = [axs]  # Make iterable if only one year

for i, year in enumerate(years):
    axs[i].bar(range(15), yearly_results[year]['lagged_corrs'])
    axs[i].set_title(f"Lagged Pearson Correlation for Year {year}")
    axs[i].set_xlabel("Lag (days)")
    axs[i].set_ylabel("Correlation Coefficient")
    axs[i].set_ylim(-1, 1)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Generate sample time series data
np.random.seed(42)
dates = pd.date_range(start="2022-01-01", periods=1000, freq="D")  # extended for multiple years
wastewater_viral_load = np.random.normal(loc=100, scale=10, size=1000).cumsum()
clinical_cases = np.roll(wastewater_viral_load, 7) + np.random.normal(loc=0, scale=50, size=1000)  # lagged + noise

df = pd.DataFrame({
    "Date": dates,
    "Wastewater_Viral_Load": wastewater_viral_load,
    "Clinical_Cases": clinical_cases
})
df.set_index("Date", inplace=True)
df['Year'] = df.index.year

# Years to analyze
years = [2022, 2023, 2024, 2025]
lags = range(0, 15)

for year in years:
    df_year = df[df['Year'] == year]
    if df_year.empty:
        print(f"No data for year {year}")
        continue

    # Calculate correlations
    pearson_corr, _ = pearsonr(df_year["Wastewater_Viral_Load"], df_year["Clinical_Cases"])
    spearman_corr, _ = spearmanr(df_year["Wastewater_Viral_Load"], df_year["Clinical_Cases"])

    # Lagged correlations
    lagged_corrs = []
    for lag in lags:
        shifted = df_year["Wastewater_Viral_Load"].shift(lag).dropna()
        aligned = df_year["Clinical_Cases"][lag:]
        corr = pearsonr(shifted, aligned)[0]
        lagged_corrs.append(corr)

    # Linear regression
    X = df_year["Wastewater_Viral_Load"].values.reshape(-1, 1)
    y = df_year["Clinical_Cases"].values
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    r2 = r2_score(y, y_pred)

    # Plot
    fig, axs = plt.subplots(3, 1, figsize=(12, 14))
    fig.suptitle(f"Analysis for Year {year}", fontsize=16)

    # Time series plot
    axs[0].plot(df_year.index, df_year["Wastewater_Viral_Load"], label="Wastewater Viral Load")
    axs[0].plot(df_year.index, df_year["Clinical_Cases"], label="Clinical Cases")
    axs[0].set_title("Wastewater Viral Load vs. Clinical COVID-19 Cases Over Time")
    axs[0].legend()
    axs[0].set_ylabel("Counts")

    # Lag correlation plot
    axs[1].bar(lags, lagged_corrs)
    axs[1].set_title("Lagged Pearson Correlation (Wastewater vs. Clinical Cases)")
    axs[1].set_xlabel("Lag (days)")
    axs[1].set_ylabel("Correlation Coefficient")
    axs[1].set_ylim(-1, 1)

    # Regression plot
    axs[2].scatter(df_year["Wastewater_Viral_Load"], df_year["Clinical_Cases"], alpha=0.6)
    axs[2].plot(df_year["Wastewater_Viral_Load"], y_pred, color="red",
                label=f"Linear Fit\nRMSE: {rmse:.2f}, R: {r2:.2f}")
    axs[2].set_title("Linear Regression: Predicting Clinical Cases from Wastewater Viral Load")
    axs[2].set_xlabel("Wastewater Viral Load")
    axs[2].set_ylabel("Clinical Cases")
    axs[2].legend()

    plt.tight_layout(rect=[0, 0, 1, 0.96])  # leave space for suptitle
    plt.show()

    # Print summary correlations
    print(f"Year {year} Summary:")
    print(f"  Pearson correlation: {pearson_corr:.3f}")
    print(f"  Spearman correlation: {spearman_corr:.3f}")
    print()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Simulated dataset to cover 20192025
np.random.seed(42)
dates = pd.date_range(start="2019-01-01", periods=2500, freq="D")
wastewater_viral_load = np.random.normal(loc=100, scale=10, size=2500).cumsum()
clinical_cases = np.roll(wastewater_viral_load, 7) + np.random.normal(loc=0, scale=50, size=2500)

df = pd.DataFrame({
    "Date": dates,
    "Wastewater_Viral_Load": wastewater_viral_load,
    "Clinical_Cases": clinical_cases
})
df.set_index("Date", inplace=True)
df['Year'] = df.index.year

# Years to analyze
years = [2019, 2020, 2021, 2022, 2023, 2024, 2025]
lags = range(0, 15)

for year in years:
    df_year = df[df['Year'] == year]
    if df_year.empty:
        print(f"No data for year {year}")
        continue

    # Correlations
    pearson_corr, _ = pearsonr(df_year["Wastewater_Viral_Load"], df_year["Clinical_Cases"])
    spearman_corr, _ = spearmanr(df_year["Wastewater_Viral_Load"], df_year["Clinical_Cases"])

    # Lagged correlations
    lagged_corrs = []
    for lag in lags:
        shifted = df_year["Wastewater_Viral_Load"].shift(lag).dropna()
        aligned = df_year["Clinical_Cases"][lag:]
        if len(shifted) == len(aligned):
            corr = pearsonr(shifted, aligned)[0]
        else:
            corr = np.nan
        lagged_corrs.append(corr)

    # Regression
    X = df_year["Wastewater_Viral_Load"].values.reshape(-1, 1)
    y = df_year["Clinical_Cases"].values
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    r2 = r2_score(y, y_pred)

    # Plotting
    fig, axs = plt.subplots(3, 1, figsize=(12, 14))
    fig.suptitle(f"Analysis for Year {year}", fontsize=16)

    # Time series
    axs[0].plot(df_year.index, df_year["Wastewater_Viral_Load"], label="Wastewater Viral Load")
    axs[0].plot(df_year.index, df_year["Clinical_Cases"], label="Clinical Cases")
    axs[0].set_title("Wastewater Viral Load vs. Clinical COVID-19 Cases Over Time")
    axs[0].legend()
    axs[0].set_ylabel("Counts")

    # Lagged correlation
    axs[1].bar(lags, lagged_corrs)
    axs[1].set_title("Lagged Pearson Correlation (Wastewater vs. Clinical Cases)")
    axs[1].set_xlabel("Lag (days)")
    axs[1].set_ylabel("Correlation Coefficient")
    axs[1].set_ylim(-1, 1)

    # Regression plot
    axs[2].scatter(df_year["Wastewater_Viral_Load"], df_year["Clinical_Cases"], alpha=0.6)
    axs[2].plot(df_year["Wastewater_Viral_Load"], y_pred, color="red",
                label=f"Linear Fit\nRMSE: {rmse:.2f}, R: {r2:.2f}")
    axs[2].set_title("Linear Regression: Predicting Clinical Cases from Wastewater Viral Load")
    axs[2].set_xlabel("Wastewater Viral Load")
    axs[2].set_ylabel("Clinical Cases")
    axs[2].legend()

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

    # Print correlation summary
    print(f"Year {year} Summary:")
    print(f"  Pearson correlation: {pearson_corr:.3f}")
    print(f"  Spearman correlation: {spearman_corr:.3f}")
    print()
```
